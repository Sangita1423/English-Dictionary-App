```python

while True:

    print('''Main Menu:  
    1. Add a new word
    2. Find the meaning
    3. Update a word
    4. Exit 
    ''')
    
    choice = input('Enter the choice:  ')
    
    import json
    import os

    
    if choice == '1':
        word = input('Enter the words: ')
        meaning = input('Enter the Meaning: ')
        if os.path.isfile('words.txt'):
            
            f = open('words.txt','r')
            temp = json.load(f)
            temp[word] = meaning
            f.close()
            
            f = open('words.txt','w')
            json.dump(temp,f)
            f.close()
            
        else:
            f = open('words.txt','w')
            dict1 = { word : meaning }
            json.dump(dict1,f)
            f.close()
            
    elif choice == '2':
        
        word = input('Enter the word:')
        
        if os.path.isfile('words.txt'):
            f = open('words.txt','r')
            temp = json.load(f)
            
            if word in temp:
                print('meaning',temp[word])
            else:
                print('The word doesnt exit is dictionary. ',temp[word])
        else:
            print('Dictionary is not created yet .')
    
    elif choice == '3':
        word = input('Enter the words: ')
        meaning = input('Enter the meaning to be updated:')
        if os.path.isfile('words.txt'):
            
            f = open('words.txt','r')
            temp = json.load(f)
            f.close()
            
            if word in temp:
                temp[word] = meaning
                f = open('words.txt','w')
                json.dump(temp,f)
                f.close()
            else:
                print('The word does not exits in the dictionary')

        else:
            print('Dictionary is not yet created.')
            
    elif choice == '4':
        break
    
    else:
        print('Invalid Choice.')
```
